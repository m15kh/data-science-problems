```python
dataset = pd.read_csv("products.csv", header=None)


transactions = []

for i in range(len(dataset)):
    # Create a temporary list to store values for the current row
    temp_row = []
    # Loop through the columns in the row
    for j in range(len(dataset.columns)):
        # Check if the value is NaN
        if not pd.isna(dataset.values[i, j]):
            # Append the non-NaN value to the temporary row list
            temp_row.append(str(dataset.values[i, j]))
    # Append the non-NaN values for the current row to the transactions list
    transactions.append(temp_row)

# Print the transactions list
transactions
```

```python
transactions1 = [
    record[1]["Product"].tolist()
    for record in list(df.groupby(["Customer Id", "Date"]))
]
transactions1
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder = TransactionEncoder()
onehot = encoder.fit_transform(transactions1)
transactions_df = pd.DataFrame(onehot, columns=encoder.columns_)
print(transactions_df)
```

```python
# # Use apriori to find frequent itemsets
frequent_itemsets = apriori(transactions_df, min_support=0.01, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0)

# Filter rules based on additional criteria if needed
# rules = rules[
#     (rules["lift"] >= 3)
#     & (rules["antecedents"].apply(len) == 2)
#     & (rules["consequents"].apply(len) == 1)
# ]

# Display the association rules
rules
```

```python
rules = apriori(
    transactions=transactions,
    min_support=0.01,  # 0.044
    min_confidence=0.7,
    min_lift=3,
    min_length=5,
    max_length=5,
)
results = list(rules)
results
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))


resultsinDataFrame = pd.DataFrame(
    inspect(results),
    columns=["Left Hand Side", "Right Hand Side", "Support", "Confidence", "Lift"],
)
```

```python
resultsinDataFrame.nlargest(n=10, columns="Lift")
```

```python
# Assuming your data is stored in a DataFrame named df
# Replace df with the name of your DataFrame if different

# Create DataFrame
df_product = pd.DataFrame(df)

# Find the maximum number of splits
split_products = df_product["Product"].str.split(" ", expand=True)

# Split 'Product' column into separate columns

# Save the split products to a CSV file
split_products.to_csv("products.csv", index=False, header=None)
```

```python
dataset = pd.read_csv("products.csv", header=None)


transactions = []

for i in range(len(dataset)):
    # Create a temporary list to store values for the current row
    temp_row = []
    # Loop through the columns in the row
    for j in range(len(dataset.columns)):
        # Check if the value is NaN
        if not pd.isna(dataset.values[i, j]):
            # Append the non-NaN value to the temporary row list
            temp_row.append(str(dataset.values[i, j]))
    # Append the non-NaN values for the current row to the transactions list
    transactions.append(temp_row)

# Print the transactions list
print(transactions)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder = TransactionEncoder()

# One-hot encode the transactions
onehot = encoder.fit_transform(transactions)

# Convert one-hot encoded data into DataFrame
transactions_df = pd.DataFrame(onehot, columns=encoder.columns_)
print(transactions_df)
```

```python
product_support = transactions_df.mean()

# Sort product support in descending order
sorted_product_support = product_support.sort_values(ascending=False)

# Print support for each product
print("Support for each product:")
print(sorted_product_support)
```

```python
# Assuming you have your transactions data stored in a DataFrame named 'transactions'

# # Use apriori to find frequent itemsets
frequent_itemsets = apriori(transactions_df, min_support=0.5, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence")

# Filter rules based on additional criteria if needed
# rules = rules[
#     (rules["lift"] >= 3)
#     & (rules["antecedents"].apply(len) == 2)
#     & (rules["consequents"].apply(len) == 1)
# ]

# Display the association rules
print(rules)
```

```python
rules = apriori(
    transactions=transactions,
    min_support=0.01,  # 0.044
    min_confidence=0.7,
    min_lift=3,
    min_length=5,
    max_length=5,
)
results = list(rules)
results
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))


resultsinDataFrame = pd.DataFrame(
    inspect(results),
    columns=["Left Hand Side", "Right Hand Side", "Support", "Confidence", "Lift"],
)
```
